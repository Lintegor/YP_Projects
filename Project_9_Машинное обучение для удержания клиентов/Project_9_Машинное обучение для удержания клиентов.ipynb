# Анализ и подготовка плана действий по удержанию клиентов фитнес клуба

## Описание проекта

Сеть фитнес-центров «Культурист-датасаентист» разрабатывает стратегию взаимодействия с клиентами на основе аналитических данных. Необходимо провести анализ и подготовить план действий по удержанию клиент. Чтобы бороться с оттоком, отдел по работе с клиентами «Культуриста-датасаентиста» перевёл в электронный вид множество клиентских анкет.

### Задачи проекта

1) спрогнозировать вероятность оттока (на уровне следующего месяца) для каждого клиента;    
2) сформировать типичные портреты клиентов - выделить несколько наиболее ярких групп и охарактеризовать их основные свойства;   
3) проанализировать основные признаки, наиболее сильно влияющие на отток;    
4) сформулировать основные выводы и разработать рекомендации по повышению качества работы с клиентами:    
   - выделить целевые группы клиентов;    
   - предложить меры по снижению оттока;    
   - определить другие особенности взаимодействия с клиентами.

### Описание данных

Заказчик подготовил данные, которые содержат данные на месяц до оттока и факт оттока на определённый месяц. Набор данных включает следующие поля:
1)	Данные клиента за предыдущий до проверки факта оттока месяц:
- **gender** — пол;
- **Near_Location** — проживание или работа в районе, где находится фитнес-центр;
- **Partner** — сотрудник компании-партнёра клуба (сотрудничество с компаниями, чьи сотрудники могут получать скидки на абонемент  в таком случае фитнес-центр хранит информацию о работодателе клиента);
- **Promo_friends** — факт первоначальной записи в рамках акции «приведи друга» (использовал промо-код от знакомого при оплате первого абонемента);
- **Phone** — наличие контактного телефона;
- **Age** — возраст;
- **Lifetime** — время с момента первого обращения в фитнес-центр (в месяцах).

2)	Информация на основе журнала посещений, покупок и информация о текущем статусе абонемента клиента:
- **Contract_period** — длительность текущего действующего абонемента (месяц, 6 месяцев, год);
- **Month_to_end_contract** — срок до окончания текущего действующего абонемента (в месяцах);
- **Group_visits** — факт посещения групповых занятий;
- **Avg_class_frequency_total** — средняя частота посещений в неделю за все время с начала действия абонемента;
- **Avg_class_frequency_current_month** — средняя частота посещений в неделю за предыдущий месяц;
- **Avg_additional_charges_total** — суммарная выручка от других услуг фитнес-центра: кафе, спорттовары, косметический и массажный салон.

3)	**Churn** — факт оттока в текущем месяце.

## Построение модели бинарной классификации клиентов

Учитывая,что у нас имеются 4ре мультиколлинеарные признака и два, кооторые совсем невлияют на признак оттока, удалим их из нашего датафейма.

```python
gym_churn.drop(['month_to_end_contract','avg_class_frequency_current_month','phone','gender'], \
               axis = 1, inplace= True)
```

```python
# разделим наши данные на признаки (матрица X) и целевую переменную (y)
X = gym_churn.drop(columns=['churn'])
y = gym_churn['churn']
```

```python
# разделяем модель на обучающую и валидационную выборку 80/20
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

Учитывая, что мы собираемся строить модель бинарной классификации, типом "Обучение с учителем" будем использовать Логистическую регрессию и 'случайный лес' без стандартизации признаков, потому что стандартизация используется для моделей, которые чувствительны к расстоянию(например кластеризация)

```python
# задайте список моделей
models = [
    LogisticRegression(random_state=0),
    RandomForestClassifier(random_state=0)]
```

```python
# функция, которая принимает на вход модель и данные и выводит метрики
def make_prediction(m, X_train, y_train, X_test, y_test):
    model = m
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(
        'Accuracy_score:{:.2f}\nPrecision_score:{:.2f}\nRecall_score:{:.2f}\n'\
         .format(
            mt.accuracy_score(y_test, y_pred),
            mt.precision_score(y_test, y_pred),
            mt.recall_score(y_test, y_pred)
        )
    )
```

```python
# напишите цикл, который выводит метрики по списку моделей
for i in models:
    print('Метрики для',i)
    make_prediction(i, X_train, y_train, X_test, y_test)
```

Доля правильных прогнозов и полнота чуть выше в модели RandomForestClassifier(Случайный лес)

```python
# обучим финальную модель
final_model = RandomForestClassifier(random_state=0)
final_model.fit(X_train, y_train)
y_pred = final_model.predict(X_test)
```

```python
# создадим датафрейм с именами признаков и их важностью и выведем его по убыванию важности
coef = abs(final_model.feature_importances_)
features = pd.DataFrame({'feature':X.columns, 'coef':coef.flatten()})
print(features.sort_values('coef', ascending=False))
```

### Вывод

Таким образом модель **RandomForestClassifier(Случайный лес)** показала наилучшие метрики:    
- Доля правильных ответов  - 89 %
- Точность предсказаний - 78 %
- Полнота предсказаний - 77 %    

Наибольшее влияние(33,4 %) на целевую переменную оказывает признак **lifetime**(время с момента первого обращения в фитнес-центр (в месяцах))

### Определение кол-ва кластеров

```python
#Удалим столбец с целевой переменной
gym_churn_cl = gym_churn.drop(columns=['churn'])
#Изза больших потребляемых мощностей вычислительной машины при построении дендрограммы
#сначала построим дендрограмму на случайной подвыборке, а после оценки оптимального числа кластеров 
#запустить более быстрый алгоритм K-Means
gym_churn_cl_part = gym_churn_cl.sample(int(gym_churn_cl.shape[0]/3)) #Берём 1/3 выборки
```

```python
#Выполним стандартизацию, перед клатеризацией
#Создадим объект класса StandardScaler и применим его ко всей выборке
gym_churn_st = StandardScaler().fit_transform(gym_churn_cl)
#Для построения дендрограммы по подвыборки
gym_churn_st_part = StandardScaler().fit_transform(gym_churn_cl_part)
```

Построим матрицу расстояний, на стандартизованной матрице признаков и нарисуем дендрограмму. На основании полученного графика определим, какое количество кластеров можно выделить.

```python
linked = linkage(gym_churn_st_part, method = 'ward')
```

```python
plt.figure(figsize=(16, 9))  
dendrogram(linked, orientation = 'top')
plt.title('Кластеризация пользователей')
plt.show()
```

Судя по дендрограмме в нашем датафрейме явно выделяется 4 кластера посетителей, но для обучения модели K-Means возьмём 5 ластеров, как написано в ТЗ

```python
gym_churn_km = KMeans(n_clusters = 5, random_state = 0)
labels = gym_churn_km.fit_predict(gym_churn_st)
gym_churn['cluster'] = labels
```
