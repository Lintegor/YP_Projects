## Описание проекта

Перед нами данные до 2016 года. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Нам нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.    
Для этого мы:    
- выполним предобработку данных;     
- добавим столбец суммарных продаж по всем регионам;    
- проведём исследовательский анализ данных, который включает:    
  - определение платформы с наибольшими суммарными продажами;
  - определение актуального периода для анализа и прогноза;
  - влияние отзывов на продажи внутри популярной платформы;
  - распределение игр по жанрам;
- составим портрет пользователя каждого региона;
- проверим гипотезы:
  - cредние пользовательские рейтинги платформ Xbox One и PC одинаковые;
  - cредние пользовательские рейтинги жанров Action и Sports разные;
- сделаем общий вывод суммирующий результаты всех исследований.  

В дата сете представлены следующие данные:
- **Name** — название игры
- **Platform** — платформа
- **Year_of_Release** — год выпуска
- **Genre** — жанр игры
- **NA_sales** — продажи в Северной Америке (миллионы проданных копий)
- **EU_sales** — продажи в Европе (миллионы проданных копий)
- **JP_sales** — продажи в Японии (миллионы проданных копий)
- **Other_sales** — продажи в других странах (миллионы проданных копий)
- **Critic_Score** — оценка критиков (максимум 100)
- **User_Score** — оценка пользователей (максимум 10)
- **Rating** — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

## Предобработка данных

```python
#приведём все столбца к нижнему регистру
data_games.columns = data_games.columns.str.lower()
```

```python
#Столбец Name
data_games[data_games['name'].isna()]
```

Отсутствие наименования игры врядли можно чем то заменить. Если бы было много пропущенных значений в столбце **name** то пропуски заполнили бы словом **отсутствует** и тем не менее могли бы использовать данные для анализа, но учитывая, что таких строк всего 2, склоняюсь к их удалению.

```python
data_games = data_games.dropna(subset = ['name'])
```

```python
#Столбец year_of_release
print(data_games['year_of_release'].isna().sum())
print('Доля пропущенных значений ', round(data_games['year_of_release'].isna().sum()/data_games.shape[0],3))
data_games[data_games['year_of_release'].isna()].sample(5)
```

Часть пропущенных значений можно заполнить годом выпуска указанным в наименовании игры. Так же, учитывая, что некоторые игры выпускались на разных платформах и на некоторых платформах годы всётаки проставлены года, их так же можно использовать. Но учитывая, что доля пропусков мала, не будем терять на это время и заполним пропущенные значения нулём. Так же переведём тип данных столбца в **int**.

```python
#Старый код V1
#data_games['year_of_release'] = data_games['year_of_release'].fillna(0).astype('int')
#Новый код V1
data_games = data_games.dropna(subset = ['year_of_release'])
data_games.year_of_release = data_games.year_of_release.astype('int')
```

После удаления строк с пропущенными значениями в наименовании игр, так же удалились строки с наименованием жанров этих игр, поэтому столбец **genre** не требует предобработки. Столбцы **critic_score, user_score** не будем заполнять значениями, потому что в дальнейшем, при построении распределений, эти медианные значения формируют не естественный столб. Тем не менее, в **user_score** необходимо убрать из столбца **tbd**. Сокращение **tbd** расшифровывается как "to be determined" в переводе означает "будет определено". Заменим эти значения на NaN и поменяем тип данных на float.

```python
#Столбец critic_score
#data_games['critic_score'].fillna(data_games['critic_score'].median(),inplace = True)
#Столбец user_score
data_games['user_score'].replace("tbd", np.NaN, inplace = True)
#data_games['user_score'].fillna(data_games['user_score'].median(),inplace = True)
data_games['user_score'] = data_games['user_score'].astype('float')
```

Столбец **rating** оставим без заполния, так как значения в нём являются категориальными, а значит средними или медианными значениями мы не можем его заполнить, да и его заполнение может повлиять на статистику.

### Изменения продаж по платформам

Для построения графика изменений продаж по каждой платформе в зависимости от года создадим цикл, который вытащит все данные по лидерам продаж из нашего исходного датафрейма. Для этого сначала определим лидеров продаж и создадимиз них список, что бы цикл мог по ним пройти.

```python
#Переменная, хранящая список из 6 лидеров продаж
top_platform = data_games.groupby('platform')['total_sales'].sum().sort_values(ascending = False)[:6].index.tolist()
#Переменная, которая будет содержать данные нашиз лидеров продаж
data_top_platform = pd.DataFrame(columns = data_games.columns.tolist())
print('Список платформ с наибольшей суммой продаж -', top_platform)
```

```python
#Наполняем наш новый датафрейм данными из исходного по лдерам продаж
for name_of_platform in top_platform:
    data_top_platform = data_top_platform.append(data_games[data_games['platform'] == name_of_platform])
#Группируем таблицу по нужным нам для графикам параметрам
data_top_platform = data_top_platform.pivot_table(index = ['platform','year_of_release'], values = 'total_sales',\
                                                 aggfunc = 'sum').reset_index()
data_top_platform
```

Строим графики продаж в зависимости от года по каждой платформе

```python
sns.lineplot(x='year_of_release', y = 'total_sales', \
             data = data_top_platform[data_top_platform['year_of_release']>1990], hue = 'platform')\
             .set(xlabel = 'Год', ylabel = 'Сумма(млн)');
```

Или с помощью баров.

```python
sns.barplot(x='year_of_release', y = 'total_sales', \
           data = data_top_platform[data_top_platform['year_of_release']>1990], hue = 'platform')
plt.xticks(rotation = 70);
```

### Самые популярные платформы ТОП-5

```python
#Функция замены значений столбца датафрейма, которые не входят в ТОП-5
def other(variable, top):
    if variable not in top:
        return 'other'
    else:
        return variable
```

```python
#Формируем списки ТОП-5 платформ по каждому региону, что бы их потом передать, как аргумент функции other
top_platform_na = data_games.groupby('platform')['na_sales'].sum().sort_values(ascending = False)[:5].index.tolist()
top_platform_eu = data_games.groupby('platform')['eu_sales'].sum().sort_values(ascending = False)[:5].index.tolist()
top_platform_jp = data_games.groupby('platform')['jp_sales'].sum().sort_values(ascending = False)[:5].index.tolist()
#Формируем столбцы, которые будут содержать только значения ТОП-5 платформ каждого региона, 
#а остальные значения будут названы как other
data_games['top_platform_na'] = data_games.apply(lambda x: other(x.platform, top_platform_na), axis =1)
data_games['top_platform_eu'] = data_games.apply(lambda x: other(x.platform, top_platform_eu), axis =1)
data_games['top_platform_jp'] = data_games.apply(lambda x: other(x.platform, top_platform_jp), axis =1)
```

```python
fig, axes = plt.subplots(1,3, figsize = (20,5))
fig.suptitle('Самые популярные платформы ТОП-5 в регионах')
axes[0].set_title('NA')
axes[1].set_title('EU')
axes[2].set_title('JP')
#Риегион NA
data_games.groupby('top_platform_na')['na_sales'].sum().plot(ax = axes[0], kind = 'pie', autopct='%1.1f%%')
#Риегион EU
data_games.groupby('top_platform_eu')['eu_sales'].sum().plot(ax = axes[1], kind = 'pie', autopct='%1.1f%%')
#Риегион JP
data_games.groupby('top_platform_jp')['jp_sales'].sum().plot(ax = axes[2], kind = 'pie', autopct='%1.1f%%');
```

## Проверка Гипотез

### Гипотеза № 1

Первая гипотеза, которую мы будем проверять, звучит следующим образом - Средние пользовательские рейтинги платформ **Xbox One** и **PC** одинаковые.    
Тогда нулевая гипотеза Ho - Средние пользовательские рейтинги платформ **Xbox One** и **PC** одинаковые.    
Противоположная гипотеза H1 - Средние пользовательские рейтинги платформ **Xbox One** и **PC** разные.    
Уровень значимости **alpha** выбираем 0.05.    
Для выбора метода проверки гепотизы узнаем размеры наших выборок, а затем при необходимости тип распределения.

```python
data_PC = data_games[data_games['platform']=='PC']['user_score'].dropna()
print('Колличество значений для платформы PC ', data_PC.shape[0])
```

```python
data_XOne = data_games[data_games['platform']=='XOne']['user_score'].dropna()
print('Колличество значений для платформы XboxOne ', data_XOne.shape[0])
```

Для проверки гипотезы можем применить ttest. Количество данных разное, поэтому **equal_var=False**.

```python
alpha = 0.05

print('Среднее значение оценок пользователей для платформы XOne', data_XOne.mean())
print('Среднее значение оценок пользователей для платформы PC', data_PC.mean())

result = st.ttest_ind(data_XOne, data_PC, equal_var = False)

print('Значение P-value ', result.pvalue )

if result.pvalue < alpha:
    print('Отвергаем гипотезу H0, в пользу гипотезы H1')
else:
    print('Не можем отвергнуть гипотезу H0')
```

```python
# Комментарий ревьюера
unique, counts = np.unique(data_XOne, return_counts=True)
a = dict(zip(unique, counts))
score = 6.8
print('Количество оценок',score, '-', a[score])
```

#### Вывод

В результате проверки гипотезы использую t-критерий Стьюдента, мы пришли к выводу, что мы не можем отвергнуть гипотезу о том, что средние оценки пользователей генеральной совокупности для платформ XOne и PC одинаковы, так как вероятность этого больше уровня значимости. 

### Гипотеза № 2

Вторая гипотеза, которую мы будем проверять, звучит следующим образом - Средние пользовательские рейтинги жанров **Action** и **Sports** разные.    
Тогда нулевая гипотеза Ho - Средние пользовательские рейтинги жанров **Action** и **Sports** одинаковые.    
Противоположная гипотеза H1 - редние пользовательские рейтинги жанров **Action** и **Sports** разные.    
Уровень значимости **alpha** выбираем 0.05.    
Для выбора метода проверки гепотизы узнаем размеры наших выборок, а затем при необходимости тип распределения.

```python
data_action = data_games[data_games['genre']=='Action']['user_score'].dropna()
print('Колличество значений для жанра Action', data_action.shape[0])
```

```python
data_sports = data_games[data_games['genre']=='Sports']['user_score'].dropna()
print('Колличество значений для жанра Sports', data_sports.shape[0])
```

Для проверки гипотезы можем применить ttest. Количество данных разное, поэтому **equal_var=False**.

```python
print('Среднее значение пользовательской оценки для жанра Action', data_action.mean())
print('Среднее значение пользовательской оценки для жанра Sports', data_sports.mean())

result = st.ttest_ind(data_action, data_sports, equal_var = False)

print('Значение P-value = ', result.pvalue)

if result.pvalue < alpha:
    print('Отвергаем гипотезу H0, в пользу гипотезы H1')
else:
    print('Не можем отвергнуть гипотезу H0')
```

#### Вывод

Мы отвергаем гипотезу о том, что средние пользовательских оценок для жанров **Action** и **Sports** одинаковы в генеральной совокупности, потому что вероятность этого меньше уровня значимости. Что означает, что средние пользовательских оценок для жанров **Action** и **Sports** в генеральной совокупности разные.
